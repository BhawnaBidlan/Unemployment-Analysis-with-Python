```python
#2019 Unemployment Rate
plt.figure(figsize=(7,7))
plt.title('Pre-Pandemic Unemployment Rate in India')
plt.barh(data['State'],data['Estimated.Unemployment.Rate'])
plt.xlabel('Estimated.Unemployment.Rate')
plt.ylabel('State')
plt.show()
```

```python
#2019 employment rate
plt.figure(figsize=(7,7))
plt.title('Pre-Pandemic Employment Rate in India ')
plt.barh(data['State'],data['Estimated.Employed'])
plt.xlabel('Estimated.Employed')
plt.ylabel('State')
plt.show()
```

```python
# 2019Sample data creation
data = pd.DataFrame({
    'State': data['State'],
    'Estimated.Unemployment.Rate': data['Estimated.Unemployment.Rate'],
    'Area': data['Area']
})

# Aggregate data: calculate mean unemployment rate for each state and area
agg_data = data.groupby(['State', 'Area'])['Estimated.Unemployment.Rate'].mean().reset_index()

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot grouped bar chart using sns.barplot
graph1=sns.barplot(x='State', y='Estimated.Unemployment.Rate', hue='Area', data=agg_data, dodge=True, ax=ax)
current_labels = [item.get_text() for item in graph1.get_xticklabels()]
graph1.set_xticklabels(current_labels, rotation=90)

# Set the labels and title
ax.set_xlabel('State')
ax.set_ylabel('Estimated Unemployment Rate')
ax.set_title('Grouped Bar Chart of Estimated Unemployment Rate by State and Area')

# Display the plot
plt.show()
```
